```python
def findMedian(arr):
    # Write your code here
    arr.sort()
    middle = len(arr)/2
    if middle % 2 == 0:
        median = (arr[int(middle-1)] + arr[int(middle)])/2
    else:
        median = arr[int(round(middle - 0.5,0))]
    return median
```

```python
def diagonalDifference(arr):
    # Write your code here
    top_down = 0
    down_up = 0
    for i in range(len(arr)):
        top_down += arr[i][i]
        down_up += arr[int(len(arr) -1 - i )][i]
    #print(top_down)
    #print(down_up)
    return abs(top_down - down_up)
```

```python
def findZigZagSequence(a, n):
    a.sort()
    mid = int((n + 1)/2 -1)
    a[mid], a[n-1] = a[n-1], a[mid]

    st = mid + 1
    ed = n - 2
    while(st <= ed):
        a[st], a[ed] = a[ed], a[st]
        st = st + 1
        ed = ed - 1

    for i in range (n):
        if i == n-1:
            print(a[i])
        else:
            print(a[i], end = ' ')
    return
```

```python
def caesarCipher(s, k):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    u_alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    h = []
    
    for i in range(len(s)):
        for j in range(len(alphabet)):
            if s[i] == alphabet[j]: 
                h.append(alphabet[(j+k)%26])
            elif s[i] == u_alphabet[j]:
                h.append(u_alphabet[(j+k)%26])
        if len(h) != i + 1:
            h.append(s[i])

    empty = [h[0]]
    for k in range(1,len(h)):
        empty[0] = empty[0]+h[k]
        
    return empty[0]
```
